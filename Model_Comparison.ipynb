# Model Comparison
Run the three model notebooks first. This notebook reads saved metrics from lr_outputs, sarima_outputs and rf_outputs and compares them.

```python
import pandas as pd, os
paths = ['lr_outputs/metrics_lr.csv','sarima_outputs/metrics_sarima.csv','rf_outputs/metrics_rf.csv']
rows = []
for p in paths:
    if os.path.exists(p):
        df = pd.read_csv(p)
        for _, r in df.iterrows():
            rows.append([p.split('/')[0], r['Target'], r['MAE'], r['RMSE'], r['R2']])
    else:
        print('Missing', p)
comp = pd.DataFrame(rows, columns=['Model','Target','MAE','RMSE','R2'])
display(comp)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
sns.barplot(data=comp, x='Model', y='R2', hue='Target')
plt.title('R2 Comparison')
plt.show()
```
